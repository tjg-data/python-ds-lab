# Series로 1차원 배열 다루기

```python
# 경고 메시지 무시
import warnings
warnings.filterwarnings(action='ignore') 

import pandas as pd
import numpy as np
```

### 1. Series 
- 판다스의 1차원 배열 구조를 가짐

```python
s1 = pd.Series([1, 2, 3])
s1
```

```python
array1 = np.array([1, 2, 3, 4, 5])
s1 = pd.Series(array1, index=['a', 'b', 'c', 'd', 'e'])
s1
```

```python
dict1 = {"number" : [1, 2, 3]}
s1 = pd.Series(dict1)
s1
```

```python
dict2 = {
    'apple': '🍎',
    'orange': '🍅',
    'lemon': '🍋'
        }
s1 = pd.Series(dict2)
print(s1)
print(s1.index)
```
