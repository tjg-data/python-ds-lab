- 특정 기능을 수행하는 코드들의 블록
- 내장 함수도 지원 : 예) print(), type()...

## 2. 함수의 형태

- 인자도 반환값도 없는 함수
    ```python
        def func1():
            <코드 블록>
    ```
  
- 인자는 있고, 반환값이 없는 함수
    ```python
        def func2(a):
            <코드 블록>
    ```

- 인자도 있고, 반환값도 있는 함수
    ```python
        def func3(a,b):
            <코드 블록>
            return sum
    ```
  

#### (1) 인자도 반환값도 없는 함수

````python
# 1. 함수 생성
def func1():
    print('first function~!')
````

````python
print(dir())
````

````python
type(func1)
````

````python
# 2. 함수 호출
func1()
````

#### (2) 인자값은 있고, 반환값이 없는 함수

````python
# 1. 함수 생성 => 호출 시 name, age 입력받아 출력
def func2(name, age):
    print(f"이름은 {name}이고, 나이는 {age}살 입니다.")
````

````python
print(dir())
````

````python
# 2. func2 호출
func2('홍길동', 30)
func2('김유신', 130)
````

#### (3) 인자값도 있고, 반환값도 있는 함수

````python
# 1. 함수 생성 => 숫자 2개를 입력받아 합계를 반환
def func3(a, b):
    return a + b
````

````python
print(dir())
````

````python
# 2. 함수 호출
func3(10, 20)
````

````python
print(f"func3(10,20) = {func3(10,20)}")
````

- 구구단 함수 생성 및 호출

````python
def gugudan(dan):
    for k in range(1, 10):
        print(f"{dan} ✖ {k} = {dan*k}")
````

````python
gugudan(3)
````

````python
def gugudan_multi(start, end):
    for i in range(1, 10):
        for j in range(start, end+1):
            print(f"{j} ✖ {i} = {i*j}", end="\t")
        print()
````

````python
gugudan_multi(5, 9)
````

- 간단한 사칙연산 계산기

````python
def calc(a, b, operator):
    while True:
        if a > b or operator == '+':
            if operator == '+':
                return a + b
            elif operator == "-":
                return a - b
        else:
            print("a값은 b값보다 커야 합니다.")
        a = int(input('a값 입력:'))
        b = int(input('b값 입력:'))
````

````python
calc(2, 8, '+')
````

````python
calc(2, 8, '-')
````

````python
calc(1, 3, '-')
````
