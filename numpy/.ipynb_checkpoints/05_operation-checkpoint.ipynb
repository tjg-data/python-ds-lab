# 배열 연산
- 배열의 shape이 같으면 데이터별로 연산이 진행됨

```python
import numpy as np
```

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
print(arr1.shape, arr2.shape)
print(arr1, arr2)
```

```python
print(arr1 + arr2)
print(arr1 - arr2)
print(arr1 * arr2)
print(arr1 / arr2)
```

```python
# shape이 다른 경우 ValueError 발생 💥
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([4, 5, 6])
print(arr1.shape, arr2.shape)
print(arr1 + arr2)
```

- 스칼라(단일 숫자)와 연산

```python
print(arr1 + 10)
print(arr1 - 1)
print(arr1 * 3)
print(arr1 / 2)
```

- 2차원 배열 연산

```python
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6], [7, 8]])
print(arr1.shape, arr2.shape)
```

```python
print(arr1 + arr2)
print(arr1 - arr2)
print(arr1 * arr2)
print(arr1 / arr2)
```
